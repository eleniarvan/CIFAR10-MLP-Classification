```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from tensorflow.keras.datasets import cifar10
(X_train, y_train), (X_test, y_test)= cifar10.load_data()
y_train = y_train.ravel()
y_test = y_test.ravel()
print ("Μέγεθος training:", X_train.shape)
print ("Μέγεθος test:", X_test.shape)

```

```python
X_train= X_train.astype("float32")/255.0
X_test= X_test.astype("float32")/255.0
X_train= X_train.reshape((X_train.shape[0], -1))
X_test= X_test.reshape((X_test.shape[0], -1))

print ("Νέα μορφή training:", X_train.shape)
print ("Νέα μορφή test:", X_test.shape)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier, NearestCentroid
from sklearn.metrics import accuracy_score
import time

data = np.load("prepared_data.npz")
X_train = data["X_train"]
X_test  = data["X_test"]
y_train = data["y_train"]
y_test  = data["y_test"]

knn1 = KNeighborsClassifier(n_neighbors=1)
knn3 = KNeighborsClassifier(n_neighbors=3)
ncc  = NearestCentroid()

classifiers = {
    "kNN (k=1)": knn1,
    "kNN (k=3)": knn3,
    "Nearest Centroid": ncc
}

for name, clf in classifiers.items():
    print(f"\n--- {name} ---")
    start = time.time()
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    elapsed = time.time() - start
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {acc:.4f}")
    print(f"Χρόνος εκτέλεσης: {elapsed:.2f} δευτερόλεπτα")



```
